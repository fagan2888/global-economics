```python
#Explore data with pandas and present basic statistics

import requests
import pandas as pd
import csv, re, sys, os
file = '/Users/dariaulybina/Desktop/georgetown/global-economics/additional_data/data_csv/LUR_PT.csv'
df = pd.read_csv(file)
df.head()
df = df.drop('Indicator', axis=1)
df.columns = ['Index','Country','LUR_PT','Year']
df = df.drop('Index', axis=1)
df.LUR_PT = df.LUR_PT.astype(float)
rng = pd.date_range(pd.to_datetime(df.Year[0]), periods=len(df.index), freq='A')
df = df.set_index(pd.DatetimeIndex(rng))
print(df.head(10))
df.describe()

```

```python
#mean unemployment across countries
mean_unempl = df.groupby('Country').mean()
print(mean_unempl.describe())
mean_unempl.head(10)
```
